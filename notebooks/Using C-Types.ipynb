# Using CTypes

```python
import ctypes

rk = ctypes.CDLL('../libradarkit.so')
```

```python
rk.RKTestByNumber(1)
```

```python
RKIntegerToCommaStyleString = rk.RKIntegerToCommaStyleString
RKIntegerToCommaStyleString.restype = ctypes.POINTER(ctypes.c_char)
cs = RKIntegerToCommaStyleString(12345678)
ctypes.c_char_p.from_buffer(cs).value.decode('utf-8')
```

```python
import numpy as np
```

```python
d = 1.0 / 5000.0;
K = 10000
x = np.array([x for x in range(K)], dtype=np.float64) * d + 700.0
u = np.array([x for x in range(K)], dtype=np.float64) * 1000
x = x + 1.0e-3 * (np.random.random(x.shape) - 0.5)
```

```python
h_x0 = np.ones(x.shape) * np.nan
h_u0 = np.ones(x.shape) * np.nan
h_dx = np.ones(x.shape) * np.nan
t = np.ones(x.shape) * np.nan
N = 1000
b = 1.0 / N
a = 1.0 - a
u0 = u[0]
x0 = x[0]
dx = 0
for i in range(N, len(x)):
    x0 = a * x0 + b * x[i]
    #u0 = u[i - N]
    u0 = a * u0 + b * u[i]
    dx_du = (x[i] - x[i - N]) / (u[i] - u[i - N])
    dx = a * dx + b * dx_du
    h_dx[i] = dx
    h_x0[i] = x0
    h_u0[i] = u0
    t[i] = x0 + dx * (u[i] - u0)
```

```python
matplotlib.pyplot.plot(np.diff(x) - 1.0 / 5000.0)
matplotlib.pyplot.plot(np.diff(t))
matplotlib.pyplot.ylim(np.array([-1, 1], dtype=np.float32) * 1.0e-3)
```

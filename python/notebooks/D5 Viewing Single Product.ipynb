# Viewing Single Product

```python
import os
import sys
import blib
import glob
import radar
import numpy as np

blib.useTheme('dark')

srcDir = os.path.abspath("../src")
if os.path.exists(srcDir):
    sys.path.insert(0, srcDir)

import radarkit
import radarkit.chart
```

```python
files = sorted(glob.glob(os.path.expanduser('~/Downloads/data/PX*.txz')))
assert len(files) > 0, "No files found"
file = files[min(len(files) - 1, 7)]
print(f"Selected file {file} ({len(files)} files)")
```

```python
data = radar.read(file)
if np.nanmax(data["products"]["P"]) < 3.142:
    print("Converting P to degrees ...")
    data["products"]["P"] = np.rad2deg(data["products"]["P"])
array = np.stack([s.filled(np.nan) for s in data["products"].values()], axis=0)
sweep = radarkit.sweep.Sweep(array,
                             symbols=data["products"].keys(),
                             time=data["time"],
                             latitude=data["latitude"],
                             longitude=data["longitude"],
                             e=data["elevations"],
                             a=data["azimuths"],
                             r=data["ranges"],
                             )
```

```python
sweep.time.strftime(r'%Y/%m/%d %H:%M:%S UTC')
```

```python
radii = np.array([1.0, *np.arange(10, 71, 10)])
single = radarkit.chart.ChartSinglePPI(size=(512, 584), s=1.0, rmax=48.0, yoff=6, radii=radii, title="Z")
# single.fig
# single.fig.savefig("test.png")
```

```python
title = f"{sweep.time.strftime(r'%Y/%m/%d %H:%M:%S UTC')}   EL: {sweep.sweepElevation:.2f}°"
single = radarkit.chart.ChartSinglePPI(sweep, size=(512, 584), s=1.0, rmax=48.0, yoff=6, radii=radii, symbol="Z", title="Z", colorbar_title=title)
single.fig
```

```python
single.set_data(sweep, symbol="V", title="V", colorbar_title=title)
single.fig
```
